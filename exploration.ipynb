### What is the average length of the loan in the dataset?

```python
# start with a standard-scaled plot
binsize = 5
bins = np.arange(0, df['Term'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = df, x = 'Term', bins = bins)
plt.xlabel('The length of the loan (months)')
plt.show()
```

Looking at the distribution of the Term of the loan, it's quite obvious that there are only 3 unique values. Lets invistigate this insight further.

```python
df['Term'].value_counts()
```

There are only 3 values for the length of the loan, a year (12 months), 3 years (36 months) or 5 years (60 months). Most of the debts in this dataset are 3 years long debts.

### What is the average Borrower's interest rate in the dataset?


```python
binsize = 0.01
bins = np.arange(0, df['BorrowerRate'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = df, x = 'BorrowerRate', bins = bins)
plt.xlabel("The Borrower's interest rate for this loan.")
plt.axvline(df['BorrowerRate'].mean(), color='orange', linestyle='-.', linewidth=1)
plt.show()
```

In the case of Borrowers's interest rate, the distribution looks roughly normal but with an isolated peak at nearly 0.32. The bulk of the interest rate is between 0.12 and 0.2 and the bulk other than the peak looks right skewed.

## Bivariate Exploration
To start off with, I want to look at the pairwise correlations present between features in the data.

```python
numeric_vars = ['year', 'Term', 'BorrowerRate', 'AvailableBankcardCredit', 'TotalProsperLoans', 'LoanOriginalAmount']
categoric_vars = ['IsBorrowerHomeowner', 'EmploymentStatus']
```

### Heatmap

```python
plt.figure(figsize = [8, 5])
sb.heatmap(df[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0)
plt.show()
```

From the heatmap above, there is no strong **linear** relation between any of our numeric variables. However, there is a weak positive linear relationship between the original amount of the loan and the term of the loan. There's also a weak negative relationship between the available bank card credit and the borrower rate which makes sense.

```python
# plot matrix: sample 500 diamonds so that plots are clearer and
# they render faster
samples = np.random.choice(df.shape[0], 500, replace = False)
df_samp = df.loc[samples,:]

g = sb.PairGrid(data = df_samp, vars = numeric_vars)
g = g.map_diag(plt.hist, bins = 20);
g.map_offdiag(plt.scatter)
```

* As expected, the shorter term loans was typically smaller (original amount), and people who chose short term loans had no prior listings or just one listing. Most of the short term debts were placed between 2011 and 2014.
* The largest loans (in amount) were placed after 2012. 
* The largest loans (in amount) didn't have a high borrower rate. It's rather the small listings that had high borrower rates.
* There is a slight trend between Borrower rate and Total prosper loans. People who took more loans tended to get loans with higher borrower rates.

## Multivariate Exploration
The main thing I want to explore in this part of the analysis is how the two categorical measures play into the relationship between LoanOriginalAmount and BorrowerRate.

```python
def hist2dgrid(x, y, **kwargs):
    palette = kwargs.pop('color')
    plt.hist2d(x, y, cmap = palette, cmin = 0.5)
```

```python
# create faceted heat maps on levels of the IsBorrowerHomeowner variable
g = sb.FacetGrid(data = df, col = 'IsBorrowerHomeowner', size = 5)
g.map(hist2dgrid, 'LoanOriginalAmount', 'BorrowerRate', color = 'inferno_r')
g.set_xlabels('LoanOriginalAmount')
g.set_ylabels('BorrowerRate')

plt.show()

```

Non-homeowners tended to get the vast majority of the small loans with high borrower rates. Their small listings took a large range of borrower rates with most of them from 0.15 to 0.35.

```python
# create faceted heat maps on levels of the IsBorrowerHomeowner variable
g = sb.FacetGrid(data = df, col = 'EmploymentStatus',col_wrap = 4 ,size = 3)
g.map(hist2dgrid, 'LoanOriginalAmount', 'BorrowerRate', color = 'inferno_r')
g.set_xlabels('LoanOriginalAmount')
g.set_ylabels('BorrowerRate')

plt.show()
```

* Interestingly, the unemployed mostly chose loans with a very high borrower rates with most of their listings having borrower rates of around 0.3 to 0.35 but with an amount of less than 10k most of the times.
* People who chose the 'Not available' option when it comes to employment status didn't get any listings with borrower rate more than 0.3 although they had a large number of listings with borrower rate of exactly 0.3 with a loan amount of less than 5k.
* Part-time workers didn't get as much high borrower rate loans as other people of other employment status. A very large amount of their loans had a borrowr rate of around 0.2.
